## Exercise 1: Protein Powder
Carbs, fats and proteins are the three primary macro nutrients. Create a figure plotting the distribution of each of these macro nutrients together (i.e., three distributions on a single plot). Make sure to provide a legend.

```python
# Enter all of your code for exercise 1 here. Feel free to add more cells if you need to:
import pandas as pd
import matplotlib.pyplot as plt
file = "cereal.csv"
df = pd.read_csv(file)
print(df)

carbs = df['carbo']
fats = df['fat']
proteins = df['protein']

plt.hist(carbs, bins=10, alpha=0.5, label='Carbohydrates')
plt.hist(fats, bins=10, alpha=0.5, label='Fats')
plt.hist(proteins, bins=10, alpha=0.5, label='Proteins')

plt.xlabel('Nutrient Value')
plt.ylabel('Frequency')
plt.legend()

plt.title('Distribution of Carbohydrates, Fats, and Proteins')

plt.show()
```

## Exercise 2: Sugar Daddy
Get a list of the top 5 most sugary cereals and the 5 least sugary cereals.

```python
# Enter all of your code for exercise 2 here. Feel free to add more cells if you need to:
sorted_sugar = df.sort_values(by='sugars', ascending=False)
most_sugar = sorted_sugar.head(5)
least_sugar = sorted_sugar.tail(5)
print("Most sugary cereals")
print(most_sugar[['name', 'sugars']])
print("Least sugary cereals")
print(least_sugar[['name', 'sugars']])
```

## Exercise 3: Cereal Killer
Get a list of the top 5 highest rated and lowest rated cereals.

```python
# Enter all of your code for exercise 3 here. Feel free to add more cells if you need to:
sorted_rating = df.sort_values(by='rating', ascending=False)
highest_rating = sorted_rating.head(5)
lowest_rating = sorted_rating.tail(5)
print("Highest rated cereals")
print(highest_rating[['name', 'rating']])
print("Lowest rated cereals")
print(lowest_rating[['name', 'rating']])
```

## Exercise 4: America
Quantify the relationship between sugar and ratings.

Make a plot to visualize this relationship. Superimpose a best fit line (with seaborn) to describe the relationship. It may be helpful to look at the [seaborn regplot documentation]("https://seaborn.pydata.org/generated/seaborn.regplot.html").

```python
# Make the plot for the data visualization and line of best fit here:
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.regplot(x='sugars', y='rating', data=df, scatter_kws={'alpha':0.5})
plt.title('Relationship Between Sugars and Ratings')
plt.xlabel('Sugars')
plt.ylabel('Rating')

plt.show()
```


Calculate a correlation statistic describing the relationship between sugar and ratings  (i.e., r or r squared).

```python
print("Correlation: ")
print(df['sugars'].corr(df['rating']))
```

Write a statement in plain English interpreting this statistic.

There is a negative correlation between the amount of sugar in the cereal. As the amount of sugar in the cereal goes up, the rating of the cereal goes down.

## Exercise 5: America Part 2
Make five plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.scatter(df['carbo'], df['rating'], alpha=0.5)
plt.xlabel('Carbo')
plt.ylabel('Rating')
plt.title('Carbo vs. Rating')

plt.subplot(2, 3, 2)
plt.scatter(df['sugars'], df['rating'], alpha=0.5)
plt.xlabel('Sugars')
plt.ylabel('Rating')
plt.title('Sugars vs. Rating')

plt.subplot(2, 3, 3)
plt.scatter(df['calories'], df['rating'], alpha=0.5)
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.title('Calories vs. Rating')

plt.subplot(2, 3, 4)
plt.scatter(df['protein'], df['rating'], alpha=0.5)
plt.xlabel('Protein')
plt.ylabel('Rating')
plt.title('Protein vs. Rating')

plt.subplot(2, 3, 5)
plt.scatter(df['fat'], df['rating'], alpha=0.5)
plt.xlabel('Fat')
plt.ylabel('Rating')
plt.title('Fat vs. Rating')

plt.show()
```

Of the variables carbo, sugars, calories, protein, and fat, which has the strongest relationship with rating? Justify your answer.

```python
print(df['carbo'].corr(df['rating']))
print(df['sugars'].corr(df['rating']))
print(df['calories'].corr(df['rating']))
print(df['protein'].corr(df['rating']))
print(df['fat'].corr(df['rating']))
```

The variable that has the strongest relationship with rating is sugars. The relationship is negative and is the closest to 1 or -1 out of all the correlations.
